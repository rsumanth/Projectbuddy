```python
#set the default color palette of a Seaborn plot using set_palette()function
def sinplot(flip=1):
    x = np.linspace(0, 15, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
        sns.set_style("white")
        sns.set_palette("husl")
        sinplot()
        plt.show()

```

```python
#Seaborn Facetgrid() function enable us to draw multi-plot grid for plotting conditional relationships.
g = sns.FacetGrid(assign, col="PLACE", row="DESCRIPTION")
g = g.map(plt.hist, "DATE")
plt.show()

```

```python
#Seaborn PairGrid() function draws subplot grid for plotting pairwise relationships in a dataset
g = sns.PairGrid(assign)
g = g.map(plt.scatter)
plt.show()

```

```python
##Seaborn PairGrid() function draws subplot grid for plotting pairwise relationships in a dataset
g = sns.PairGrid(assign)
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)

```
